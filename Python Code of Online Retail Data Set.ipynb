```python
import pandas as pd

url = "https://raw.githubusercontent.com/Salwa-L123/Online-Retail-Data-Set/main/Online%20Retail%20Data%20Set.xlsx"
df = pd.read_excel(url, engine='openpyxl')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df['InvoiceNo'] = df['InvoiceNo'].astype(str).str.replace(r'[A-Za-z]', '', regex=True)
df['StockCode'] = df['StockCode'].astype(str).str.replace(r'[A-Za-z]', '', regex=True)
```

```python
df.head(157)
```

```python
before = len(df)
df = df.dropna(subset=['Description'])
after = len(df)

print(f"Deleted {before - after} rows with missing product descriptions 💨")
```

```python
df = df[(df['Quantity'] >= 1) & (df['UnitPrice'] > 0)]
df.head(7)
```

```python
df['CustomerID'] = df['CustomerID'].fillna(0).astype(int)  # Fill NaNs with 0 before converting
```

```python
df.head()
```

```python
df = df.dropna(subset=['CustomerID'])
df['CustomerID'] = df['CustomerID'].astype(int)  # just to make sure it’s clean
```

```python
df.to_csv('cleaned_data.csv', index=False)
```
